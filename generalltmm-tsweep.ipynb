define constants & load AM1.5

```python
q=1.602e-19
c=2.998e8
h=6.626e-34

am1p5_raw=np.loadtxt('am1p5.txt')
```

load nk data - as a dictionary with the different materials as items, arrays are lambda (microns), n, k

```python
nkdata_raw = np.load('nkdata_raw.npy',allow_pickle='TRUE').item()
print(nkdata_raw.keys())
```

## calculate absorbtion in each layer of a stack

### > input parameters here <
initialize stack parameters

```python
# list of materials in stack
mat_list=np.array(['SiO_2','ITO','PTAA','WS_2','C60','BCP','Ag'])

# list of layer thicknesses in nm, start and end with inf
d_list = np.array([np.inf,1e6,100,8,20,2,2,100,np.inf])

# list of wavelengths
ldas=np.arange(350,850,1) 

#polarization
pol="s"

#angle
theta=0;

#coherent or incoherent
coherent=1 #1 for coherent, 0 for incoherent
if coherent==0:
    c_list=['c']*len(d_list) #default
    c_list[0]='i' #first and last layer incoherent
    c_list[-1]='i' #first and last layer incoherent
    c_list[np.where(np.logical_and(d_list<np.inf,d_list>1e3))[0][0]]='i' #only sets first thick layer to incoherent
```

```python
#find which layer TMD is in
true_array=np.logical_or(np.logical_or(mat_list=="MoS_2",mat_list=='MoSe_2'),
              np.logical_or(mat_list=='WS_2',mat_list=='WSe_2'))

tmd_index=np.argwhere(true_array==True)[0][0]
c60_index=np.argwhere(mat_list=='C60')[0][0]

c60_ds=np.arange(0,10,1)
tmd_ds=np.arange(10,30,1)

c60_ds
```

### transfer matrix stuff

interpolate nk data

```python
nkdata={}
for mat in mat_list:
    raw=nkdata_raw[mat+"_raw"]
    n_raw=interp1d(raw[:,0],raw[:,1],fill_value="extrapolate")
    k_raw=interp1d(raw[:,0],raw[:,2],fill_value="extrapolate")
    n=n_raw(ldas/1000)
    k=k_raw(ldas/1000)
    nkdata[mat]=np.column_stack((ldas,n,k))

#interpolate am1.5
am1p5_int=interp1d(am1p5_raw[:,0],am1p5_raw[:,1],fill_value="extrapolate")
am1p5=am1p5_int(ldas)
```

actual transfer matrix calculation

```python
#initialize all matrices
R=np.zeros(np.size(ldas))
T=np.zeros(np.size(ldas))
A=np.zeros(np.size(ldas))
a_layer=np.zeros((np.size(ldas),np.size(mat_list)))
jscs=np.zeros((np.size(c60_ds),np.size(tmd_ds)))

#loop over C60 thicknesses
i_c60=0
for c60_d in c60_ds:
    
    d_list[c60_index+1]=c60_d

    #loop over TMD thicknesses
    i_tmd=0
    for tmd_d in tmd_ds:
        
        d_list[tmd_index+1]=tmd_d
        
        #loop over all wavelengths
        i=0
        for lda in ldas:    

            #initialize n_list, index
            n_list=np.ones(np.size(d_list),dtype=complex)
            j=0

            #calculate n,k at this wavelength for each material, populate n_list
            for mat in mat_list:
                j+=1
                nk=nkdata[mat]
                n_list[j]=nk[i,1]+nk[i,2]*1j

            if coherent==1:
                #TMM
                tmm_data=coh_tmm(pol,n_list,d_list,theta,lda)

                #absorption in each layer
                a=absorp_in_each_layer(tmm_data)

            elif coherent==0:
                #TMM
                tmm_data=inc_tmm(pol,n_list,d_list,c_list,theta,lda)

                #absorption in each layer
                a=inc_absorp_in_each_layer(tmm_data)

            #add to arrays 
            R[i]=tmm_data["R"]
            T[i]=tmm_data["T"]
            A[i]=1-tmm_data["R"]-tmm_data["T"]
            a_layer[i,:]=a[1:np.size(mat_list)+1]

            i+=1

        #calculate jsc
        integrand=am1p5*a_layer[:,tmd_index]*q/(h*c)*ldas*1e-9
        jsc=trapz(integrand,ldas)/10
        jscs[i_c60,i_tmd]=jsc
        
        i_tmd+=1
    
    i_c60+=1
```

visualize Jsc

```python
#C60 = rows = x 
#TMD = columns = y 

y,x=np.meshgrid(tmd_ds,c60_ds)

fig,ax=plt.subplots()

cplot=ax.pcolor(x,y,jscs)
cbar=fig.colorbar(cplot)
cbar.set_label('$J_{SC}$ $(mA/cm^2)$')#, rotation=270)

ax.set_xlabel('C60 Thickness (nm)')
ax.set_ylabel('TMD Thickness (nm)')
#ax.set_title('ITO, 8 nm PTAA, 5 nm BCP, Ag')

fig.savefig('cscs/C60_TMD_tsweep_inverted_WS2.png')
```
