## calculate absorbtion in each layer of a stack

### > input parameters here <
initialize stack parameters

```python
#parameters to play with
t_ws2=40
ends=2
delta_ws2_1=0.5-1j
delta_ws2_2=0.5-1j
delta_ws2_3=0.5-1j

# list of materials in stack
mat_list=np.array(["SiO_2","Pt","WS_2","WS_2","WS_2","Ag"]) #assume oil index matches to SiO2

# list of layer thicknesses in nm, start and end with inf
d_list = np.array([np.inf,50,10,ends,t_ws2-2*ends,ends,100,np.inf])
tmd_d_list = np.arange(0,20,10)

# excitation wavelength
lda=635

#polarization
pol="s"

#angle
theta=0;
```

### transfer matrix stuff

interpolate nk data

```python
nkdata={}
for mat in mat_list:
    raw=nkdata_raw[mat+"_raw"]
    n_raw=interp1d(raw[:,0],raw[:,1],fill_value="extrapolate")
    k_raw=interp1d(raw[:,0],raw[:,2],fill_value="extrapolate")
    n=n_raw(lda/1000)
    k=k_raw(lda/1000)
    nkdata[mat]=np.column_stack((lda,n,k))
```

reflected: determine phase of Ef and Eb across entire stack

```python
#initialize n_list, index
n_list=np.ones(np.size(d_list),dtype=complex)
index=0

#calculate n,k at this wavelength for each material, populate n_list
for mat in mat_list:
    index+=1
    nk=nkdata[mat]
    n_list[index]=nk[0,1]+nk[0,2]*1j

n_list[0]=n_list[1] #oil index-matched to silica

#TMM
coh_tmm_data=coh_tmm(pol,n_list,d_list,theta,lda)

#calculate phase within ws2 
vw_list=coh_tmm_data["vw_list"]
kz_list=coh_tmm_data["kz_list"]

ef_R=np.zeros(int(np.sum(d_list[1:-1])),dtype=complex)
eb_R=np.zeros(int(np.sum(d_list[1:-1])),dtype=complex)
phasef_R=np.zeros(int(np.sum(d_list[1:-1])))
phaseb_R=np.zeros(int(np.sum(d_list[1:-1])))
abs_R=np.zeros(int(np.sum(d_list[1:-1])))

index=0
for mat in mat_list:
    index+=1

    for z in np.arange(0,d_list[index],1):
        z_tot=int((np.cumsum(d_list[1:-1])-d_list[1:-1])[index-1]+z)
        delta_n=z*kz_list[index]
        ef_R[z_tot]=vw_list[index,0]*np.exp(delta_n*1j)
        eb_R[z_tot]=vw_list[index,1]*np.exp(-delta_n*1j)
        phasef_R[z_tot]=np.angle(ef_R[z_tot])
        phaseb_R[z_tot]=np.angle(eb_R[z_tot])

        test=position_resolved(index,z,coh_tmm_data)["Ey"]
        abs_R[z_tot]=position_resolved(index,z,coh_tmm_data)["absor"]

        if test!=ef_R[z_tot]+eb_R[z_tot]:
            print('does not agree with TMM Ey')
```

scattered: now change WS2 index slightly & repeat

```python
#initialize n_list, index
n_list=np.ones(np.size(d_list),dtype=complex)
index=0

#calculate n,k at this wavelength for each material, populate n_list
for mat in mat_list:
    index+=1
    nk=nkdata[mat]
    n_list[index]=nk[0,1]+nk[0,2]*1j

n_list[0]=n_list[1] #oil index-matched to silica
n_list[3]=n_list[3]+delta_ws2_1
n_list[4]=n_list[4]+delta_ws2_2
n_list[5]=n_list[5]+delta_ws2_3


#TMM
coh_tmm_data=coh_tmm(pol,n_list,d_list,theta,lda)

#calculate phase within ws2 
vw_list=coh_tmm_data["vw_list"]
kz_list=coh_tmm_data["kz_list"]

ef_S=np.zeros(int(np.sum(d_list[1:-1])),dtype=complex)
eb_S=np.zeros(int(np.sum(d_list[1:-1])),dtype=complex)
phasef_S=np.zeros(int(np.sum(d_list[1:-1])))
phaseb_S=np.zeros(int(np.sum(d_list[1:-1])))

index=0
for mat in mat_list:
    index+=1

    for z in np.arange(0,d_list[index],1):
        z_tot=int((np.cumsum(d_list[1:-1])-d_list[1:-1])[index-1]+z)
        delta_n=z*kz_list[index]
        ef_S[z_tot]=vw_list[index,0]*np.exp(delta_n*1j)
        eb_S[z_tot]=vw_list[index,1]*np.exp(-delta_n*1j)
        phasef_S[z_tot]=np.angle(ef_S[z_tot])
        phaseb_S[z_tot]=np.angle(eb_S[z_tot])

        test=position_resolved(index,z,coh_tmm_data)["Ey"]

        if test!=ef_S[z_tot]+eb_S[z_tot]:
            print('does not agree with TMM Ey')
```

    #calculate fresnel coefficients 
    vw_list=coh_tmm_data["vw_list"]
    
    kz_list=coh_tmm_data["kz_list"]
    delta_n=d_list[1:-1]*kz_list[1:-1]
    vw_list[1:-1,0]=vw_list[1:-1,0]*np.exp(delta_n*1j) #propagate v and w to the end of each layer 
    vw_list[1:-1,1]=vw_list[1:-1,1]*np.exp(delta_n*-1j) #v will decrease, w will increase (absorption)
    vw_list[0]=[1,coh_tmm_data["r"]]
    vw_list[-1]=[coh_tmm_data["t"],0]
        
    phaseflip = np.logical_or(
                    np.logical_and((vw_list[:,0] > 0),(vw_list[:,1] < 0)), 
                    np.logical_and((vw_list[:,1] > 0),(vw_list[:,0] < 0))
                                            )
        
    vw_amp=np.square(np.absolute(vw_list))
    max_backwards=np.max(vw_amp[2:,1])
    
    rmax=np.where(vw_amp==np.max(vw_amp[2:,1]))
    interface=np.asscalar(rmax[0])
    
    phaseflips[str(d)+'nm']=phaseflip
    
    if interface==2:
        interfaces[str(d)+'nm']=2 #reflection from Pt/WS2
    elif interface==3:
        interfaces[str(d)+'nm']=3 #reflection from WS2/Ag
    else:
        print('other')
